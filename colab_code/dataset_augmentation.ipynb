```python
# Rot Down 

from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img

datagen = ImageDataGenerator(
        rotation_range=5,
        fill_mode='nearest')

for x in range (200):
  img = load_img('/content/push up dataset/pushups_down/down ('+ str(x+1) + ').jpg')  # this is a PIL image
  x = img_to_array(img)  # this is a Numpy array with shape (3, 150, 150)
  x = x.reshape((1,) + x.shape)  # this is a Numpy array with shape (1, 3, 150, 150)

  # the .flow() command below generates batches of randomly transformed images
  # and saves the results to the `preview/` directory
  i = 0

  for batch in datagen.flow(x, batch_size=32,save_to_dir='/content/push up dataset/pushups_down/', save_prefix='down_rot', save_format='jpg'):
    i += 1
    if i > 0:
        break  # otherwise the generator would loop indefinitely
```

```python
# Flip Down

from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img

datagen = ImageDataGenerator(
        horizontal_flip=True,
        fill_mode='nearest')

for x in range (100):
  img = load_img('/content/push up dataset/pushups_down/down ('+ str(x+1) + ').jpg')  # this is a PIL image
  x = img_to_array(img)  # this is a Numpy array with shape (3, 150, 150)
  x = x.reshape((1,) + x.shape)  # this is a Numpy array with shape (1, 3, 150, 150)

  # the .flow() command below generates batches of randomly transformed images
  # and saves the results to the `preview/` directory
  i = 0

  for batch in datagen.flow(x, batch_size=32,save_to_dir='/content/push up dataset/pushups_down/', save_prefix='down_flip', save_format='jpg'):
    i += 1
    if i > 0:
        break  # otherwise the generator would loop indefinitely
```
